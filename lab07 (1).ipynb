Діедральна група $D_{20}$ має ряд нормальних підгруп: 

```python
G = DihedralGroup(20)
H = DihedralGroup(20)
norm_subs = G.normal_subgroups()

print("D20")
gens = G.gens()
phi = PermutationGroupMorphism(G, H, [H(()), H(())])
ker_D20 = phi.kernel()
print(ker_D20)
print(ker_D20 == norm_subs[0])
print("ker = ", ker_D20.structure_description())

print("==========================")

print("C20")
gens = G.gens()
phi = PermutationGroupMorphism(G, H, [H(()), gens[1]])
ker_C20 = phi.kernel()
print(ker_C20)
print(ker_C20 == norm_subs[1])
print("ker = ", ker_C20.structure_description())

print("==========================")

print("D10_1")
gens = G.gens()
phi = PermutationGroupMorphism(G, H, [gens[1], gens[1]])
ker_D10_1 = phi.kernel()
print(ker_D10_1)
print(ker_D10_1 == norm_subs[3])
print("ker = ", ker_D10_1.structure_description())

print("==========================")

print("D10_2")
gens = G.gens()
phi = PermutationGroupMorphism(G, H, [gens[0]^10, H(())])
ker_D10_2 = phi.kernel()
print(ker_D10_2)
print(ker_D10_2 == norm_subs[2])
print("ker = ", ker_D10_2.structure_description())

print("==========================")

print("C10")
gens = G.gens()
phi = PermutationGroupMorphism(G, H, [gens[0]^10, gens[1]])
ker_C10 = phi.kernel()
print(ker_C10)
print(ker_C10 == norm_subs[4])
print("ker = ", ker_C10.structure_description())

print("==========================")

print("C5")
gens = G.gens()
phi = PermutationGroupMorphism(G, H, [gens[0]^15, gens[1]^3])
ker_C5 = phi.kernel()
print(ker_C5)
print(ker_C5 == norm_subs[7])
print("ker = ", ker_C5.structure_description())

print("==========================")

print("C4")
gens = G.gens()
phi = PermutationGroupMorphism(G, H, [gens[0]^4, gens[0]^9 * gens[1]])
ker_C4 = phi.kernel()
print(ker_C4)
print(ker_C4 == norm_subs[5])
print("ker = ", ker_C4.structure_description())

print("==========================")

print("C2")
gens = G.gens()
phi = PermutationGroupMorphism(G, H, [gens[0]^2, gens[1]^3])
ker_C2 = phi.kernel()
print(ker_C2)
print(ker_C2 == norm_subs[6])
print("ker = ", ker_C2.structure_description())

print("==========================")

print("nothing")
gens = G.gens()
phi = PermutationGroupMorphism(G, H, [gens[0], gens[1]])
ker_none = phi.kernel()
print(ker_none)
print(ker_none == norm_subs[8])
print("ker = ", ker_none.structure_description())
```

```python
   groups = {
    "C11": CyclicPermutationGroup(11),
    "D13": DihedralGroup(13),
    "Q8": QuaternionGroup(),
    "SL(3,Z7)": SL(3, GF(7))
}

results = {}
for name, G in groups.items():
    center = G.center()
    conjugacy_classes = G.conjugacy_classes()
    num_elements = G.order()
    num_classes = len(conjugacy_classes)
    class_representatives = [cl.representative() for cl in conjugacy_classes]
    results[name] = {
        "center": center.list(),
        "num_elements": num_elements,
        "num_classes": num_classes,
        "class_representatives": class_representatives
    }

results
```
